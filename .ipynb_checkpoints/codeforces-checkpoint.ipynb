```python
# utils


def get_contest_data():
    url = "https://codeforces.com/api/contest.list"
    response = requests.get(url)
    data = json.loads(response.text)

    if data["status"] == "OK":
        contests = data["result"]

        for contest in contests:
            print("Contest ID:", contest["id"])
            print("Contest Name:", contest["name"])
            print("Type:", contest["type"])
            print("Phase:", contest["phase"])
            print("=============================================")
    else:
        print("Failed to retrieve contest data. Reason:", data["comment"])


def format_problem_statement(problem_statement):
    # Define the formatting elements
    header_prefix = "Problem Statement:"
    indentation = " " * 4
    section_headers = ["Task:", "Example:", "Constraints:"]

    # Replace formatted number with bold format
    problem_statement = problem_statement.replace("$$$", "**")
    problem_statement = problem_statement.replace("**", "**", 1)

    # Split the problem statement into sections
    sections = problem_statement.split("\n\n")

    # Format each section
    formatted_sections = []
    for section in sections:
        formatted_section = f"{indentation}{section}"
        formatted_sections.append(formatted_section)

    # Combine the formatted sections with headers and indentation
    formatted_string = f"{header_prefix}\n\n"
    formatted_string += "\n\n".join(formatted_sections)

    # Add section headers
    for header in section_headers:
        formatted_string += f"\n\n{header}"

    return formatted_string


def get_problem_data(tags=None):
    url = "https://codeforces.com/api/problemset.problems"

    if tags:
        url += "?tags=" + tags

    response = requests.get(url)
    data = json.loads(response.text)

    if data["status"] == "OK":
        problems = data["result"]["problems"]

        for problem in problems:
            print("Problem Name:", problem["name"])
            print("Contest ID:", problem["contestId"])
            print("Index:", problem["index"])
            print("Type:", problem["type"])
            print("Tags:", ", ".join(problem["tags"]))
            print("=============================================")
    else:
        print("Failed to retrieve problem data. Reason:", data["comment"])


def get_problem_statement(contest_id, problem_index):
    url = f"https://codeforces.com/problemset/problem/{contest_id}/{problem_index}"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    problem_statement = soup.find("div", class_="problem-statement")

    if problem_statement:
        statement = problem_statement.text
        print(statement)

        conversation = [
            {
                "role": "system",
                "content": f"You are given the following codeforces problem statement:{statement}.You need to output proper python dict in response to user query about this problem statement.",
            },
            {
                "role": "user",
                "content": "what are possible edge cases in this problem?",
            },
        ]
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo-0613",
            messages=conversation,
            max_tokens=1000,
            temperature=0,
            n=1,
        )
        if "choices" in response:
            model_response = response["choices"][0]["message"]["content"]
            print("Model Response:")
            print(model_response)
        else:
            print("Failed to generate model response.")
    else:
        print("Failed to retrieve problem statement.")
```
