```python
import requests
from bs4 import BeautifulSoup

url = "https://leetcode.com/problems/two-sum/"

# Send a GET request to the URL
response = requests.get(url)

# Create a BeautifulSoup object with the response text
soup = BeautifulSoup(response.text, "html.parser")

# Find the div tag with class "_1l1MA" and data attribute "description_content"
problem_div = soup.find(
    "div", class_="_1l1MA", attrs={"data-track-load": "description_content"}
)

# Extract the problem data from the div tag
problem_data = problem_div.get_text()

# Print the problem data
print(problem_data)
```
